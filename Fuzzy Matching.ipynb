```python
df_b_clean = df_b[["date_time", "bill_number", "reff_num", "transaction_made"]]
df_b_clean = df_b_clean.dropna()
df_b_clean.describe()
```

```python
df_b_clean = df_b_clean.rename(columns={"date_time": "TRANS_DATE", "bill_number": "BILL_NUMBER", "reff_num": "REFF_NUMBER", "transaction_made": "TRANS_AMOUNT"})
df_b_clean.info()
```

```python
# create a new dataframe, with 2 columns, first columns is still date, second column is combination of bill number + ref number + transaction
df_a_combined = pd.DataFrame({'DATE': df_a_clean['TRANS_DATE'], 'DATA': df_a_clean['BILL_NUMBER'].astype(str) + df_a_clean['REFF_NUMBER'].astype(str) + df_a_clean['TRANS_AMOUNT'].astype(str)})
df_a_combined.head()
```

```python
df_b_combined = pd.DataFrame({'DATE': df_b_clean['TRANS_DATE'], 'DATA': df_b_clean['BILL_NUMBER'].astype(str) + df_b_clean['REFF_NUMBER'].astype(str) + df_b_clean['TRANS_AMOUNT'].astype(str)})
df_b_combined.head()
```

```python
merged_df = pd.merge(df_a_combined, df_b_combined, on='DATA', how='inner')
```

```python
new_df = merged_df.rename(columns={"DATE_x": "DATE_A", "DATE_y": "DATE_B"})
# rearrange the columns
new_df = new_df[['DATA', 'DATE_A', 'DATE_B']]
new_df.head()
```

```python
# count NA values
new_df.isna().sum()
```

```python
new_df.info()
```

```python
new_df
```

```python
import numpy as np
new_df['DATE_A'], new_df['DATE_B'] = np.where(new_df['DATE_B'] < new_df['DATE_A'], 
                                       (new_df['DATE_B'], new_df['DATE_A']), 
                                       (new_df['DATE_A'], new_df['DATE_B']))
news_df = new_df[['DATA','DATE_A','DATE_B']]
new_df['DATA'] = news_df['DATA']

news_df['TIME_DIFF'] = new_df['DATE_B'] - new_df['DATE_A']

news_df
```

```python
import numpy as np
news_df['DATE_A'], news_df['DATE_B'] = np.where(news_df['DATE_B'] < news_df['DATE_A'], 
                                       (news_df['DATE_B'], news_df['DATE_A']), 
                                       (news_df['DATE_A'], news_df['DATE_B']))

# Menghitung perbedaan waktu dalam detik
news_df['TIME_DIFF'] = (news_df['DATE_B'] - news_df['DATE_A']).dt.total_seconds()

# Menghitung nilai maksimum perbedaan waktu
max_diff = news_df['TIME_DIFF'].max()

# Menghindari pembagian dengan nol
max_diff = max(max_diff, 1)  # Pastikan max_diff tidak nol

# Menghitung persentase kemiripan berdasarkan perbedaan waktu
news_df['SIMILARITY_PERCENTAGE'] = (1 - (news_df['TIME_DIFF'] / max_diff)) * 100

# Menampilkan hasil
news_df[['DATA', 'DATE_A', 'DATE_B', 'TIME_DIFF', 'SIMILARITY_PERCENTAGE']]
```

```python
countS_df = news_df.groupby('TIME_DIFF').size().reset_index(name='count')

countS_df
```

```python
news_df[['DATE_A','DATE_B']] = news_df[['DATE_A','DATE_B']].apply(pd.to_datetime, errors='coerce')
news_df['DATA'] = news_df['DATA'].astype(str)
final_df = news_df[['DATA','DATE_A','DATE_B']]
final_df
```

```python
# change all to string
final_df = final_df.astype(str)
```

```python
# export
final_df.to_excel("output.xlsx", index=False)
```

```python
from fuzzywuzzy import fuzz
def calculate_similarity(str1, str2):
    return fuzz.ratio(str1, str2)
def calculate_similarity_percentage(df):
    df['similarity_percentage'] = df.apply(
        lambda row: calculate_similarity(row['DATE_A'], row['DATE_B']), axis=1
    )
    return df

ffinal_df = calculate_similarity_percentage(final_df)

ffinal_df
```

```python
count_df = ffinal_df.groupby('similarity_percentage').size().reset_index(name='count')

count_df
```
